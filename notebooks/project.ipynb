# 第一階段：資料整理與觀察

```python
# 下載 CityC Challenge Data
import os
import gdown

if not os.path.exists("../data/CityC Challenge Data.csv.gz"):
    FILE_ID = "1kOy1GyjtCDWRg0skF-i5H2ywSFi_qddm"
    gdown.download(id=FILE_ID, output="../data/CityC Challenge Data.csv.gz", quiet=False)
else:
    print("File already exists.")
```

```python
# 解壓縮
import gzip
import shutil

gz_file_path = "../data/CityC Challenge Data.csv.gz"
output_file_path = "../data/sapporo_challengedata.csv"

if os.path.exists(output_file_path):
    print("File already exists.")
else:
    with gzip.open(gz_file_path, "rb") as f_in:
        with open(output_file_path, "wb") as f_out:
            shutil.copyfileobj(f_in, f_out)
    print(f"已解壓縮 {gz_file_path} 到 {output_file_path}")
```
